# Feature Engineering

1. Create New Feature(salary_in_usd)

```python
data['salary_per_remote'] = data['salary_in_usd'] / (data['remote_ratio'] + 1)
```

2. Binning / Categorization

          (Convert continuous values into categories).

```python
data['salary_level'] = pd.cut(
    data['salary_in_usd'],
    bins=[0, 50000, 100000, 200000, data['salary_in_usd'].max()],
    labels=['Low', 'Medium', 'High', 'Very High']
)
```

3. Grouping Data using groupby()

```python
grouped_data = data.groupby('experience_level')['salary_in_usd'].mean()

print(grouped_data)
```

4. Aggregating Data using agg()

```python
aggregated_data = data.groupby('experience_level').agg({
    'salary_in_usd': ['mean', 'max', 'min']
})

print(aggregated_data)
```

# Data Split/Partition

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,      # 20% testing data
    random_state=42     # reproducibility
)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# Data Transformation

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Model Initialization

```python
from sklearn.svm import SVR

model = SVR()
```

# Model Training

```python
model.fit(X_train, y_train)
```

# Prediction

```python
y_pred = model.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

# Model Visualization
